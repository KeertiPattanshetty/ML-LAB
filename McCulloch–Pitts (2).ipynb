```python
# AND Gate using MCP Neuron
def AND_gate(a, b):
    # Weights = [1, 1], Threshold = 2
    weighted_sum = (a * 1) + (b * 1)
    return 1 if weighted_sum >= 2 else 0

# Truth Table
print("A B | AND")
for A in [0, 1]:
    for B in [0, 1]:
        print(f"{A} {B} |  {AND_gate(A,B)}")
```

```python
# OR Gate using MCP Neuron
def OR_gate(a, b):
    # Weights = [1, 1], Threshold = 1
    weighted_sum = (a * 1) + (b * 1)
    return 1 if weighted_sum >= 1 else 0

# Truth Table
print("A B | OR")
for A in [0, 1]:
    for B in [0, 1]:
        print(f"{A} {B} |  {OR_gate(A,B)}")
```

```python
# XOR Gate using a network of MCP Neurons
def AND_gate(a, b):
    return 1 if (a + b) >= 2 else 0

def OR_gate(a, b):
    return 1 if (a + b) >= 1 else 0

def NOT_gate(a):
    return 1 if (-1 * a) >= 0 else 0

def XOR_gate(a, b):
    h1 = OR_gate(a, b)
    h2 = AND_gate(a, b)
    h3 = NOT_gate(h2)
    return AND_gate(h1, h3)

# Truth Table
print("A B | XOR")
for A in [0, 1]:
    for B in [0, 1]:
        print(f"{A} {B} |  {XOR_gate(A,B)}")
```
